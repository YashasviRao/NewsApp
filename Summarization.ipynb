### Automating the following tasks with Selenium Web-driver



*   Extracting the urls of the articles from the home page of 'nbcnews.com'
*   Removing the unnecessary urls such as of ads



```python
driver.get("https://www.nbcnews.com")
links = driver.find_elements(By.TAG_NAME,"a");
urls = set()
for link in links:
  k = link.get_attribute('href')
  k = str(k)
  #From manual observation, I found that the links related to articles contains rcna in it's url. Hence, we collect them into a list
  if k.find('rcna')!=-1:
    urls.add(k)
print(urls)
```

Performing the following tasks from the article description page of the website

*   Extracting Title
*   Extracting Image url

*   Extracting Article body
*   Ignoring the irrelevant text from the body of article





```python
temp = urls
articles = []
adSet=set()
num=0
for url in temp:
  try:
    article = []
    print(url)
    driver.get(url)
    Heading = driver.find_element(By.TAG_NAME,'header')
    title=Heading.find_element(By.TAG_NAME,'h1').text
    article.append(title)
    text = driver.find_element(By.TAG_NAME,'article')
    body = text.find_elements(By.TAG_NAME,'p')
    image = driver.find_element(By.TAG_NAME,'img')
    imageUrl=image.get_attribute('src')
    print('image :', imageUrl)
    article.append(imageUrl)
    #print(body.text)
    x = ""
    try: 
      ad = driver.find_element(By.CLASS_NAME,'nl-signup-inline')
      ad_body = ad.find_elements(By.TAG_NAME,'p')
      for content in ad_body:
        adSet.add(content.text)
    except Exception:
      pass 
    finally:
      for content in body:
        if content.text in adSet:
          continue
        else:
          x += "\n"+content.text
      article.append(x)
      articles.append(article)
      num=num+1
  except Exception:
    pass
```

### Method for building Sentence Similarity

```python
def sentence_similarity(sent1, sent2, stopwords=None):    
     if stopwords is None:        
       stopwords = []     
     sent1 = [w.lower() for w in sent1]    
     sent2 = [w.lower() for w in sent2]     
     all_words = list(set(sent1 + sent2))    
     vector1 = [0] * len(all_words)    
     vector2 = [0] * len(all_words)        
     for w in sent1:
         if w in stopwords:
             continue
         vector1[all_words.index(w)] += 1
     for w in sent2:
         if w in stopwords:
             continue        
         vector2[all_words.index(w)] += 1     
     return 1 - cosine_distance(vector1, vector2)
```

### Generating Similarity Matrix

```python
def build_similarity_matrix(sentences, stop_words):
    # Create an empty similarity matrix
    similarity_matrix = np.zeros((len(sentences), len(sentences)))
    for idx1 in range(len(sentences)):
        for idx2 in range(len(sentences)):
            if idx1 == idx2: #ignore if both are same sentences
                continue
            similarity_matrix[idx1][idx2] = sentence_similarity(sentences[idx1], sentences[idx2], stop_words)
    return similarity_matrix
```

### Final Summarization Method

```python
def nltk_summarizer(raw_text):
    raw_text=  re.sub(r'[^a-zA-z0-9.,!?/:;\"\'\s]', '', raw_text)
    stop_words = stopwords.words('english')
    summarize_text = []
    # Step 1 - Read text and tokenize
    sentences = sent_tokenize(raw_text)
    # Step 2 - Generate Similary Martix across sentences
    sentence_similarity_martix = build_similarity_matrix(sentences, stop_words)
    # Step 3 - Rank sentences in similarity martix
    sentence_similarity_graph = nx.from_numpy_array(sentence_similarity_martix)
    scores = nx.pagerank(sentence_similarity_graph)
    # Step 4 - Sort the rank and pick top sentences
    ranked_sentence = sorted(((scores[i],s) for i,s in enumerate(sentences)), reverse=True)    
    #print("Indexes of top ranked_sentence order are ", ranked_sentence)
    #print(len(raw_text))
    top_n=3
    for i in range(top_n):
        summarize_text.append("".join(ranked_sentence[i][1]))
    # Step 5 -output the summarize text
    a=(" ".join(summarize_text))
    x=Rouge()
    y=x.get_scores(a,raw_text)
    return a
```
